```python
#importing libraries left to analyze the data
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("austin_weather_final.csv")
data
```

```python
x= data.drop(['PrecipitationSumInches'], axis=1)
y= data['PrecipitationSumInches']
```

```python
#reshaping it into 2d vector format
y= y.values.reshape(-1,1)
y
```

```python
days_index= 798
days= [i for i in range(y.size)]
```

```python
#initialize the linear regression classifier
clf= LinearRegression()

#train the classifier
clf.fit(x,y)
```

```python
#plotting the graph
print('The precipitation trend graph:')
plt.scatter(days, y, color='blue')
plt.title('Precipitation level')
plt.xlabel('Days')
plt.ylabel('Precipitation in inches')
plt.scatter(days[days_index], y[days_index], color='red')
plt.show()
x_vis=x.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
print("Precipitation vs attributes trend graph:")

for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,x_vis[x_vis.columns.values[i][:100]], color='green')
    plt.scatter(days[days_index],x_vis[x_vis.columns.values[i]][days_index], color='orange')
    plt.title(x_vis.columns.values[i])
plt.show()
```

### Thus, we may conclude the use of linear regression in real life rainfall prediction.
